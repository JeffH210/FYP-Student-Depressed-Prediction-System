**Detect & Treat Outliers (IQR) (Age and CGPA)**

```python
Q1_Age = df['Age'].quantile(0.25)
Q3_Age = df['Age'].quantile(0.75)
IQR_Age = Q3_Age - Q1_Age

Q1_CGPA = df['CGPA'].quantile(0.25)
Q3_CGPA = df['CGPA'].quantile(0.75)
IQR_CGPA = Q3_CGPA - Q1_CGPA
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Import Tensorflow packages

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.regularizers import l2
```

### Sequential ANN Model
1. 128 layers (first layer) - kernel regularizer
2. 64 layers (second layer)
3. 32 layers (third layer)
4. output layer (sigmodid function) output 0 or 1 based on the final prediction probabilities

```python
model = Sequential([
    Dense(128, activation= 'relu', kernel_regularizer=l2(0.001), input_shape=(X_train.shape[1],)),
    Dropout(0.3),

    Dense(64, activation= 'relu'),
    Dropout(0.3),

    Dense(32, activation= 'relu'),
    Dropout(0.1),

    Dense(1, activation= 'sigmoid')

])
```

### Model Optimizer
- Adam
- Binary crossentropy
- accuracy

```python
# Compile model
model.compile(
    optimizer ='adam',
    loss= 'binary_crossentropy',
    metrics= ['accuracy']
)
```

```python
from keras.callbacks import ReduceLROnPlateau
```

### Import ReduceLR to reduce the learning rate based on the layers to prevent increase the bias weightage and overfitting
1. 150 epochs
2. 64 batch size
3. lr_scheduler (ReduceLR)

```python
#Train
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=10, min_lr=1e-6, verbose=1)
model.fit(X_train, y_train, epochs=150, batch_size=64, callbacks=[lr_scheduler])
```

# Fine Tuning ANN

## Using Keras Tuner 
1. Bayesian Search Optimization - finding the best model performance to evaluate on Test data
2. Number of layers selection (64, 128, 256)
3. Batch Normalization
4. Dropout rate selection (0.0， 0.3, 0.5)
5. Learning rate selection (1e-2, 1e-3, 1e-4)

```python
import keras_tuner as kt
from tensorflow.keras.layers import Dense, Input, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import regularizers
import tensorflow as tf
```

```python
def build_fine_tune_model(hp):
    tf.keras.backend.clear_session()  # safety

    model = Sequential()
    model.add(Input(shape=(X_train.shape[1],)))

    # Tune number of hidden layers
    num_layers = hp.Int('num_layers', min_value=1, max_value=6)

    for i in range(num_layers):
        neurons = hp.Choice(f'units_{i}', [64, 128, 256])
        model.add(Dense(
            neurons, 
            activation='relu'
            ))

        model.add(BatchNormalization())

        dropout_rate = hp.Choice(f'dropout_{i}', [0.0, 0.2, 0.5])
        if dropout_rate > 0:
            model.add(Dropout(dropout_rate))

    # Output layer
    model.add(Dense(1, activation='sigmoid'))

    # Learning rate
    lr = hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])

    model.compile(
        optimizer=Adam(learning_rate=lr),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model
```

```python
import os
```

* Note: os directory file name as "kt_dir" and it will store at C:Users\kt_dir (FIrst time Install Keras Tuner ANN)

```python
directory = os.path.join(os.path.expanduser("~"), "kt_dir")
os.makedirs(directory, exist_ok=True)
```

### Bayesian Optimization
- objective : finding the best validation accuracy for each trails loop
- max_trails = 25
- executions per 1 trial
- each run the Jupyter Notebook will overwrite the history (you can turn it as False)

```python
tuner = kt.BayesianOptimization(
    build_fine_tune_model,
    objective= 'val_accuracy',
    max_trials= 25,
    executions_per_trial= 1,
    directory=directory,  
    project_name= "binary_fine_tune_ann_tuning",
    overwrite = True
)
```

```python
tuner.search_space_summary()
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping

reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.7, patience=5, min_lr=1e-6, verbose=1)

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
tuner.search(
    X_train, y_train, 
    validation_split=0.2, 
    epochs=150, 
    batch_size=32,
    callbacks=[reduce_lr, early_stop],
)
```

```python
# Get the best model
best_fine_tune_ANNmodel = tuner.get_best_models(num_models=1)[0]

best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
train_loss, train_acc = best_fine_tune_ANNmodel.evaluate(X_train, y_train, verbose=0)
test_loss, test_acc = best_fine_tune_ANNmodel.evaluate(X_test, y_test, verbose=0)

print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy:  {test_acc:.4f}")
```
